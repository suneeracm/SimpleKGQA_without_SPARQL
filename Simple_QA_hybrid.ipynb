```python
import tagme
import re
# Set the authorization token for subsequent calls.
def tagmee(q):
  uris=[]
  mentions=[]
  tagme.GCUBE_TOKEN = "588fa38a-608c-45b1-8739-6300206c01f9-843339462"
  tags = tagme.annotate(q)
  prefi="http://dbpedia.org/resource/"
  pref1="https://en.wikipedia.org/wiki/"
  #Print annotations with a score higher than 0.1
  try:
    for ann in tags.get_annotations(0.1):
      mentions.append(ann.mention)
      uris.append(ann.uri())
    dben=[]
    for i in uris:
      ent = re.sub(pref1,'',i)
      dben.append(prefi+ent)
  except:
    print("Tagme Error..")

  return mentions, dben
```

```python
import spacy_dbpedia_spotlight
def spotlight_ent(query):
  ent_list=[]
  nlp =spacy_dbpedia_spotlight.create('en')
  try:
    doc=nlp(query)
    entities = [(ent.text, ent.label_,ent.kb_id_ ) for ent in doc.ents]
    #print("Entities are: ", entities)
    ent_word=""
    for ent in doc.ents:
      ent_list.append(ent.kb_id_)
      ent_word=ent.text
  except:
    print("Spotlight Error")
  return ent_list
```

```python
import requests
import csv


headers = {'content-type': 'application/json', 'Accept-Charset': 'UTF-8'}


def falcon_call(text):
    url = 'https://labs.tib.eu/falcon/api?mode=long'
    entities=[]
    relations=[]
    payload = '{"text":"'+text+'"}'
    r = requests.post(url, data=payload.encode('utf-8'), headers=headers)
    if r.status_code == 200:
        response=r.json()
        for result in response['entities']:
            try:
                entities.append(result[0])
            except:
                continue
        for result in response['relations']:
            try:
                relations.append(result[0])
            except:
                continue
    else:
        r = requests.post(url, data=payload.encode('utf-8'), headers=headers)
        if r.status_code == 200:
            response=r.json()
            for result in response['entities']:
                try:
                    entities.append(result[0])
                except:
                    continue
            for result in response['relations']:
                try:
                    relations.append(result[0])
                except:
                    continue

    return entities,relations
```
